# DeepFlowerNet : Flower Recognition Using Keras

```python
import os
import numpy as np
from keras.preprocessing import image
from keras.models import model_from_json
```

```python
class_dict = {0:'daisy', 1:'dandelion', 2:'rose', 3:'sunflower', 4:'tulip'}
```

## Loading Trained Model

```python
json_file = open('model.json', 'r')
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
model.load_weights("model.h5")
```

```python
model.summary()
```

## Image of Sunflower

```python
sunflower_path = 'E://Data Science Projects//Flower Recognition//Flowers_Data//sunflower//2721638730_34a9b7a78b.jpg'
sunflower_image = image.load_img(sunflower_path, target_size = (240, 320, 3))
```

```python
sunflower_image
```

```python
sunflower_image = image.img_to_array(sunflower_image)
sunflower_image = np.expand_dims(sunflower_image, axis=0)
predicted_class = np.argmax(model.predict(sunflower_image), axis=1)[0]

print('Actual Image Class : Sunflower')
print('Predicted Image Class : ' + str(class_dict[predicted_class]))
```
